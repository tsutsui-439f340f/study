```python
import os
import sys
import shutil
import numpy as np
```

```python
out="E:yolo_data/"
os.mkdir(out)
os.mkdir(os.path.join(out,"images"))
os.mkdir(os.path.join(out,"labels"))
```

```python
path="E:pytorch_yolov3-main/vott-csv-export/"
dir=os.listdir(path)
count=0
first=0
with open(os.path.join(path,"イラストオブジェクト検出-export.csv"),encoding="utf-8") as lines:
    f=0
    pre_name=""
    label_txt=""
    for line in lines:
        data=line.replace("\n","").replace('"','').split(",")
        name=data[0].split(".")
        if len(name)>1:
            label=data[-1]
            c=data[1:-1]
            new_file=str(count)+"."+name[1]
            
            if first==1:
                if pre_name!=data[0]:
                    #ここまでの結果を保存する
                    shutil.copy(os.path.join(path,pre_name),out+"images/"+str(count)+"."+pre_name.split(".")[1])
                    with open(os.path.join(out,"labels/{}.txt".format(str(count))),"w") as f:
                        f.write(label_txt)
                    count+=1
                    label_txt=""  
            else:
                first=1
                
            label_txt+="{} {} {} {} {}\n".format(label,c[0],c[1],c[2],c[3])
            pre_name=data[0]
           
            
n_test=count//10
n_train=count-n_test
train_idx=set(np.random.choice(count,n_train,replace=False))
test_idx=set(np.arange(count)).difference(train_idx)
with open(os.path.join(out,"train.txt"),"w") as f:
    for i in train_idx:
        f.write(str(i)+".jpg\n")
with open(os.path.join(out,"test.txt"),"w") as f:
    for i in test_idx:
        f.write(str(i)+".jpg\n") 
```
