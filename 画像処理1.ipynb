```python
img=cv2.imread("E:\material\EibiUnVVgAUQWMC.jfif")
```

```python
def show(img):
    fig, ax = plt.subplots(figsize=(16, 16))
    ax.imshow(img[:,:,[2,1,0]])
```

```python
show(img)
```

```python
def make_circle(z):
    v=np.zeros((z,z),dtype="uint8")
    v= cv2.circle(v,((z//2),(z//2)), z//2, 1,-1)
    return v
```

```python
im=img.copy()
gray=cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)
mask=np.zeros_like(gray)
z=103
p=make_circle(z)

h,w=img.shape[0:2]
for i,j in zip(np.random.randint(0,h,size=170),np.random.randint(0,w,size=170)):
    if not (i-(z//2)<0) and not (i+(z//2)>h):
        if  not (j-(z//2)<0) and not (j+(z//2)>w):
            mask[i-(z//2):i+(z//2)+1,j-(z//2):j+(z//2)+1]|=p
mask[mask==255]=1
```

```python
plt.imshow(p,"gray")
```

```python
plt.imshow(mask,"gray")
```

```python
b=np.expand_dims(mask,2)
b=b*[1,1,1]
```

```python
r=im*b
```

```python
show(r)
```
